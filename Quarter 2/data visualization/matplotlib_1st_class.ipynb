```python
from sklearn.datasets import load_iris
iris = load_iris()
display(iris)
```

```python
features=iris.data.T
plt.scatter(features[0], features[1], alpha=0.8, s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.features_names[0])
plt.ylabel(iris.features_names[1])
```

```python
x = np.linspace(1, 1000, 1000000)
y = x+2
plt.plot(x,y, 'k--')
plt.xlabel('x')
plt.ylabel('sinx')
plt.grid(True, which='major', color='b', linestyle='-')
plt.grid(True, which='minor', color='r', linestyle='--')
plt.minorticks_on()
plt.title('Graph of Sin x')
```
